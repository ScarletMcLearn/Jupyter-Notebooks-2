# All of Statistics example 3.46
## Transformations of random variables

$$X \sim \mathrm{Uniform}(-1, 4)$$
$$Y = X^2$$

What is the pdf of $Y$? That is, if we draw N samples from X and then square the results, what is the probability of each result?



```python
draws = uniform(-1, 4).rvs(100000)
plt.hist(draws, bins=100, normed=True)
plt.xlim(-2, 4)
plt.show()
plt.hist(draws ** 2, bins=100, normed=True)
plt.xlim(0, 6)
plt.show()
```
